```python
#上の二つのコードは練習（下書き）ですので、こちらを提出課題として添削をお願い致します。
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def check_adult(self):
        if self.age >= 20:
            print(f"{self.name}さんは大人です")
        else:
            print(f"{self.name}さんは大人ではありません")

human_names = {"太郎": 23, "次郎": 18, "三郎": 15}

for name, age in human_names.items():
    human = Human(name, age)
    human.check_adult()
```
